```python
import pandas as pd

train_data = pd.read_csv('train.csv')
gender_data = pd.read_csv('gender_submission.csv')
test_data = pd.read_csv('test.csv')
```

```python
train_data = train_data.dropna()
test_data = test_data.dropna()
```

```python
train_data['Sex'] = train_data['Sex'].apply(lambda x : 1 if(x=='male') else 0)
train_data['Embarked'] = train_data['Embarked'].apply(lambda x : 1 if x=='S' else 2 if x=='C' else 3 if x == 'Q' else 0 )
train_data['Embarked'].unique()
test_data['Sex'] = test_data['Sex'].apply(lambda x : 1 if(x=='male') else 0)
test_data['Embarked'] = test_data['Embarked'].apply(lambda x : 1 if x=='S' else 2 if x=='C' else 3 if x == 'Q' else 0 )
test_data['Embarked'].unique()
```

```python
from sklearn.ensemble import RandomForestRegressor
features = ['Pclass','Sex','Age','SibSp','Parch','Embarked','Fare']

X = train_data[features]
y = train_data['Survived']
test_data_input = test_data[features]
```

```python
test_data
```

```python

rfr = RandomForestRegressor(random_state = 1)
rfr.fit(X,y)
```

```python
rfr_predictions = rfr.predict(test_data_input).astype('int32')
rfr_predictions
```

```python
data = pd.DataFrame(test_data['PassengerId'],columns=['PassengerId','Survived'])
data['Survived']=rfr_predictions
data.reset_index(drop=True, inplace=True)
data.to_csv('my_predictions.csv',index=False);
data
```
